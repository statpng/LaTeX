## 1. Texlive 설치

 - [texlive 홈페이지](https://www.tug.org/texlive/) 접속
 - "install-tl.zip"과 비슷한 이름의 설치파일을 다운받고 설치하면 됨.
 
 * texlive vs miktex
  - 현재 LaTeX 문법을 지원하는 프로그램이 대표적으로 texlive와 miktex 두 가지 프로그램이 존재함. 
  - [KTUG](http://www.ktug.org/xe/) 그룹에서 예전에는 두 프로그램에 한글관련 업데이트를 진행해왔지만 현재는 texlive에서만 진행하는 듯 함.
  - miktex에서 한글 사용이 불가능한것은 아니지만 설정하기 까다로움.
  - 본인이 문서에 한글을 전혀 안 쓸 것이라면 miktex을 설치해도 무방함.

## 5. 자주 사용하는 내용
 - newcommand, def 선언

   아래와 같이 수학 기호의 경우에는 종류가 매우 다양하여 매크로보다는 newcommand/def 형태로 선언해서 사용하는 것이 좋다.
   
   (예시)
  * \newcommand{\argmin}{\mathop{\mathrm{argmin}}}
  * \def\R{\mathbb{R}}
  
 - section, subsection 적극 활용
  * section 등의 단위로 문서를 작성하면 "table of contents" 생성이나 문서의 이해력을 향상시키는 데에 도움이 된다.
 
 - \label, \ref 활용
  * 특정 Figure 또는 Table을 인용하는 경우, * Figure에 \label을 설정하고, \ref로 본문에 인용하자.
  * 예시: "Figure \ ref{fig:scatter_plot}"
  * 본문에서 "Figure 1" 등으로 숫자로 입력하게되면 Figure number가 바뀌게 되면 일일이 다 수정해야한다...
 
 - Ctrl + Alt + Up/Down : 여러 행을 선택할 수 있음.  
 
   엑셀 --> tabular 등의 테이블 생성할 때 Gooooooood!!

 - Adaboost algorithm
 
```
 \begin{frame}
	\frametitle{Boosting}
	\framesubtitle{(1) AdaBoost algorithm}
	\small{Consider a 2-class problem with $Y \in \{  -1, 1 \}$.}
	\begin{minipage}{.95 \linewidth}
	\footnotesize{
	\begin{tabular}{l}
		\toprule[.5mm]
		\textbf{Algorithm 4.1} AdaBoost algorithm.\\
		\hline
		\begin{minipage}[t]{1\textwidth}
			\vspace{-0.2cm}
			\begin{enumerate}[label=(\roman*)]  % roman, arabic, alph
				\setlength{\itemsep}{0.13cm}
				\item Start with the weights $w_i = \dfrac{1}{n}, i=1, \cdots, n$
				\item Repeat for $m=1, \cdots, M$:
				\begin{enumerate}[label=(\alph*)]
					\setlength{\itemsep}{0.15cm}
					\item Fit a weak classifier $\delta$ that minimizes the weighted error rate
					\vspace{-.2cm}
					\begin{equation*}
					\text{err} = \sum_{i=1}^n w_i I(Y_i \ne \delta (X_i))
					\vspace{-.2cm}
					\end{equation*}
					
					and call the fitted classifier $f_m$, and its corresponding error rate $\text{err}_m$.
					
					\item Compute
					\vspace{-.2cm}
					\begin{equation*}
					c_m = \log \left( \dfrac{1-\text{err}_m}{\text{err}_m} \right)
					\vspace{-.2cm}
					\end{equation*}
					
					\item Update the weights $w_i$ by
					\vspace{-.2cm}
					\begin{equation*}
					w_i = \dfrac{w_i \exp \{ c_m I(Y_i \ne f_m (X_i)) \}}{\sum_{j=1}^n w_j \exp \{  c_m I(Y_i \ne f_m (X_j)) \}}
					\vspace{-.2cm}
					\end{equation*}
				\end{enumerate}
				\item Output the classifier
				\vspace{-.3cm}
				\begin{equation*}
				f(x) = \text{sgn} \{  \sum_{m=1}^M c_m f_m(x) \}
				\end{equation*}
			\end{enumerate}
			
		\end{minipage} \\
		\hline
	\end{tabular}
	}

	\end{minipage}
\end{frame}
```

 - Spectral clustering algorithm
```
\begin{minipage}{.95\linewidth}
			\begin{enumerate}[label={(\roman*)}] % alph* roman* arabic*
			\item Unnormalized spectral clustering
			
			\begin{tabular}{ll}
				STEP 1 : & \hspace{-.4cm} Compute $\by_i , ~i=1, \cdots, n$ : $i$th row of $\bZ = (\bz_1, \cdots, \bz_k )$,\\
				
				& \hspace{-.4cm} where $\bz_1, \cdots, \bz_k $ : $k$-smallest eigenvectors of $\bL$.\\
				
				STEP 2 : & \hspace{-.4cm} Apply $k$-means clustering to $\by_1, \cdots, \by_n$.\\
			\end{tabular}
			\\
			\item Normalized spectral clustering (Ng, Jordan, and Weiss 2002)
			
			\begin{tabular}{ll}
				STEP 1 : & \hspace{-.4cm} Compute $\by_i , ~i=1, \cdots, n$ : $i$th row of $\bZ = (\bz_1, \cdots, \bz_k )$,\\
				
				& \hspace{-.4cm} where $\bz_1, \cdots, \bz_k $ : $k$-smallest eigenvectors of $\bL^* = \bD^{-1/2} \bL \bD^{-1/2}$.\\
				
				STEP 2 : & \hspace{-.4cm} Form $n \times k$ matrix $\bT$ from $\bZ$ by normalizing the rows to norm 1.\\
				
				STEP 3 : & \hspace{-.4cm} Apply $k$-means clustering to $\bt_1, \cdots, \bt_n$, where $\bt_i$ is $i$th row of $\bT$.\\
			\end{tabular}
			
		\end{enumerate}
	\end{minipage}
```
